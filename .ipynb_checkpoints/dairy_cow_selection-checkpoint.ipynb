```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

def analyzis_quantity(df, num, name, x_label):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∏—Ç "–∫–æ—Ä–æ–±–æ—á–Ω—ã–π" –≥—Ä–∞—Ñ–∏–∫ –∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É 
    –ø–æ —É–∫–∞–∑–∞–Ω–Ω–æ–º—É —Å—Ç–æ–ª–±—Ü—É –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏ –µ–≥–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –º–µ—Ç—Ä–∏–∫–∏.
    –ê—Ä–≥—É–º–µ–Ω—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏:
    df - –¥–∞–Ω–Ω—ã–µ
    num - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ "–∫–æ—Ä–∑–∏–Ω" –Ω–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ
    name - –Ω–∞–∑–≤–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
    x_label - –ø–æ–¥–ø–∏—Å—å –¥–ª—è –æ—Å–∏ –•
    '''
    # –°–æ–∑–¥–∞–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
    fig, (ax_box, ax_hist) = plt.subplots(2, sharex=True, 
                                          figsize=(7, 5), 
                                          gridspec_kw={'height_ratios': (.15, .85)})
    sns.boxplot(x=df, orient='h', ax=ax_box)
    sns.histplot(x=df, bins=num, ax=ax_hist)

    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π
    ax_box.set(xlabel='')
    fig.suptitle(name, fontsize=14)
    plt.xlabel(x_label)
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')

    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
    plt.tight_layout()
    plt.show()

    # –í—ã–≤–æ–¥ —Å—Ç–∞—Ç. –º–µ—Ç—Ä–∏–∫ —Å—Ç–æ–ª–±—Ü–∞
    print(df.describe())
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–π–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π

def analyzis_category(df, name):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–≤–æ–¥–∏—Ç —Å—Ç–æ–ª–±—á–∞—Ç—ã–π –≥—Ä–∞—Ñ–∏–∫
    –ø–æ —É–∫–∞–∑–∞–Ω–Ω–æ–º—É —Å—Ç–æ–ª–±—Ü—É –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ –∏ –µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è 
    –≤ —Ç–∞–±–ª–∏—á–Ω–æ–º –≤–∏–¥–µ.
    –ê—Ä–≥—É–º–µ–Ω—Ç—ã —Ñ—É–Ω–∫—Ü–∏–∏:
    df - –¥–∞–Ω–Ω—ã–µ
    name - –Ω–∞–∑–≤–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞
    '''
    # –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
    category_count = df.value_counts(ascending=True)

    # –°–æ–∑–¥–∞–Ω–∏–µ —Å—Ç–æ–ª–±—á–∞—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
    plot_bar = category_count.plot(kind='bar', figsize=(7, 5), grid=True) 

    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π
    plt.title(name)
    plt.xlabel('')
    plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
    plt.xticks(rotation=0)

    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
    plt.tight_layout()
    plt.show()

    # –í—ã–≤–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Ç–∞–±–ª–∏—á–Ω–æ–º –≤–∏–¥–µ
    print(category_count)
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –¥–∏–∞–≥—Ä–∞–º–º —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è

def scatter_plot(column , attribute):
    '''
    –ù–∞ –≤–≤–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–∞–µ—Ç –∫–æ–ª–æ–Ω–∫—É –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ 
    –ø–æ –∫–æ—Ç–æ—Ä–æ–π –¥–µ–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è –∏ —Å–ø–∏—Å–æ–∫ –∫–æ–ª–æ–Ω–æ–∫, 
    –∫–æ—Ç–æ—Ä—ã–µ –±—É–¥—É—Ç –Ω–∞—Ö–æ–¥–∏—Ç—å—Å—è –Ω–∞ –æ—Å–∏ x. 
    –ù–∞ –≤—ã–≤–æ–¥–µ –ø–æ–ª—É—á–∞–µ–º —Å–æ—Å—Ç–∞–≤–Ω–æ–π –≥—Ä–∞—Ñ–∏–∫ 
    —Å –¥–∏–∞–≥—Ä–∞–º–º–∞–º–∏ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è.
    '''
    # –°–æ–∑–¥–∞–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
    fig, axes = plt.subplots(3, 2, figsize=(11, 13))

    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –º–∞—Å—Å–∏–≤ –æ—Å–µ–π –≤ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π
    axes = axes.flatten()

    # –¶–∏–∫–ª –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
    for i, ax in enumerate(attribute):
        sns.scatterplot(data=df_ferma_main, x=ax, y='milk_yield_kg', 
                        hue=column, ax=axes[i])
        
        # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–∫–æ–≤ –∏ –ª–µ–≥–µ–Ω–¥—ã
        axes[i].set_title(f'–î–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è: {ax} vs milk_yield_kg')
        axes[i].legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), ncol=3, fontsize=10)

    # –°–∫—Ä—ã–≤–∞–µ–º –ø–æ—Å–ª–µ–¥–Ω—é—é (–ø—É—Å—Ç—É—é) —è—á–µ–π–∫—É
    axes[-1].axis('off')

    # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫–∏
    plt.tight_layout()
    plt.show()
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∑–Ω–∞—á–µ–Ω–∏–π
def spr_category(column):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞
    –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏—é, –∫ –∫–æ—Ç–æ—Ä–æ–π –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ
    –¥–æ 0,87 - "–Ω–∏–∑–∫–∏–π",
    –æ—Ç 0,87 –¥–æ 0,91 - "—Å—Ä–µ–¥–Ω–∏–π",
    –æ—Ç 0,91 - "–≤—ã—Å–æ–∫–∏–π"
    '''
    if column < 0.87:
        return '–Ω–∏–∑–∫–∏–π'
    elif 0.87 <= column < 0.91:
        return '—Å—Ä–µ–¥–Ω–∏–π'
    elif 0.91 <= column:
        return '–≤—ã—Å–æ–∫–∏–π'
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ –∑–Ω–∞—á–µ–Ω–∏–π
def spr_category_2(column):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞
    –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏—é, –∫ –∫–æ—Ç–æ—Ä–æ–π –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏–µ
    –¥–æ 0,91 - "–Ω–∏–∑–∫–∏–π",
    –æ—Ç 0,91 - "–≤—ã—Å–æ–∫–∏–π"
    '''
    if column < 0.91:
        return '–Ω–∏–∑–∫–∏–π'
    elif 0.91 <= column:
        return '–≤—ã—Å–æ–∫–∏–π'
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
def learning_model_lr(X, y, size, num_col_names, cat_col_names):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –æ–±—É—á–∞–µ—Ç –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏,
    –Ω–∞ –≤–≤–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–∞–µ—Ç:
    X - –≤—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏,
    y - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫,
    size - —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
    num_col_names - —Å–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, 
    cat_col_names - —Å–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –≤—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    
    –ù–∞ –≤—ã–≤–æ–¥–µ —Ñ—É–Ω–∫—Ü–∏–∏ –º—ã –ø–æ–ª—É—á–∞–µ–º:
    model_lr - –æ–±—É—á–µ–Ω–Ω—É—é –º–æ–¥–µ–ª—å,
    predictions - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –ø–æ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ,
    X_train - –æ–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞,
    X_test - —Ç–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞,
    y_train - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏,
    y_test - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
    '''
    # –î–µ–ª–∏–º –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size = size,
        random_state=RANDOM_STATE
        )
    
    # –ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
    X_test_ohe = encoder.transform(X_test[cat_col_names])

    # –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    X_train_scaled = scaler.fit_transform(X_train[num_col_names])
    X_test_scaled = scaler.transform(X_test[num_col_names])

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    encoder_col_names = encoder.get_feature_names_out()

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã
    X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
    X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

    X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

    # –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    X_train = pd.concat([X_train_scaled, X_train_ohe], axis=1)
    X_test = pd.concat([X_test_scaled, X_test_ohe], axis=1)

    # –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
    model_lr = LinearRegression()
    model_lr.fit(X_train, y_train)

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏
    predictions = model_lr.predict(X_test)
    
    return model_lr, predictions, X_train, X_test, y_train, y_test
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –∏ –∞–Ω–∞–ª–∏–∑–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤
def quality_model_lr(model_lr, y_test, predictions):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–∏—Ç—ã–≤–∞–µ—Ç –º–µ—Ç—Ä–∏–∫—É "r2_score",
    —Ä–∞—Å—á–∏—Ç—ã–≤–∞–µ—Ç –æ—Å—Ç–∞—Ç–∫–∏ –∏ —Å—Ç—Ä–æ–∏—Ç –Ω–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É —Å
    –¥–∏–∞–≥—Ä–∞–º–º–æ–π —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è –¥–ª—è –∏—Ö –∞–Ω–∞–ª–∏–∑–∞, –∏ –≤—ã–≤–æ–¥–∏—Ç
    –≤–µ—Å–∞ –º–æ–¥–µ–ª–∏ –∏ –µ–µ —Å–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç.
    
    –ù–∞ –≤–≤–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç:
    model_lr - –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –∫–æ—Ç–æ—Ä—É—é –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º
    y_test - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
    predictions - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏

    –ù–∞ –≤—ã–≤–æ–¥–µ —Ñ—É–Ω—Ü–∏–∏ –º—ã –ø–æ–ª—É—á–∞–µ–º:
    r2 - –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ "r2_score",
    residuals - –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –æ—Å—Ç–∞—Ç–∫–∏
    '''
    
    # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ r2_score
    r2 = r2_score(y_test, predictions)

    # –†–∞—Å—á–µ—Ç –æ—Å—Ç–∞—Ç–∫–æ–≤
    residuals = y_test - predictions

    # –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –æ—Å—Ç–∞—Ç–∫–æ–≤
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))

    # –ü–µ—Ä–≤—ã–π –≥—Ä–∞—Ñ–∏–∫
    axes[0].hist(residuals)
    axes[0].set_title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤')
    axes[0].set_xlabel('–û—Å—Ç–∞—Ç–∫–∏')

    # –í—Ç–æ—Ä–æ–π –≥—Ä–∞—Ñ–∏–∫
    axes[1].scatter(x=predictions, y=residuals)
    axes[1].set_title('–ê–Ω–∞–ª–∏–∑ –¥–∏—Å–ø–µ—Ä—Å–∏–∏')
    axes[1].set_xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏')
    axes[1].set_ylabel('–û—Å—Ç–∞—Ç–∫–∏')

    # –í—ã–≤–æ–¥ –≥—Ä–∞—Ñ–∏–∫–æ–≤, —Ä–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç—Ä–∏–∫–∏ –∏ –≤–µ—Å–∞ –º–æ–¥–µ–ª–∏
    plt.show()
    print(f'–†–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç—Ä–∏–∫–∏ R2: {r2:.3f}')
    print()
    print('–í–µ—Å–∞ –ø–µ—Ä–≤–æ–π –º–æ–¥–µ–ª–∏: ', model_lr.coef_)
    print('–°–≤–æ–±–æ–¥–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç: ', model_lr.intercept_)

    return r2, residuals
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
def learning_model_clf(X, y, size, num_col_names, cat_col_names):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –æ–±—É—á–∞–µ—Ç –º–æ–¥–µ–ª—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏,
    –Ω–∞ –≤–≤–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è –ø–æ–ª—É—á–∞–µ—Ç:
    X - –≤—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏,
    y - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫,
    size - —Ä–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
    num_col_names - —Å–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, 
    cat_col_names - —Å–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –≤—Ö–æ–¥–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    
    –ù–∞ –≤—ã–≤–æ–¥–µ —Ñ—É–Ω–∫—Ü–∏–∏ –º—ã –ø–æ–ª—É—á–∞–µ–º:
    model_clf - –æ–±—É—á–µ–Ω–Ω—É—é –º–æ–¥–µ–ª—å,
    predictions - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –ø–æ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ,
    X_train - –æ–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞,
    X_test - —Ç–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞,
    y_train - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏,
    y_test - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
    '''
    # –î–µ–ª–∏–º –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏
    X_train, X_test, y_train, y_test = train_test_split(
        X,
        y,
        test_size = size,
        random_state=RANDOM_STATE
        )
    
    # –ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
    X_test_ohe = encoder.transform(X_test[cat_col_names])

    # –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    X_train_scaled = scaler.fit_transform(X_train[num_col_names])
    X_test_scaled = scaler.transform(X_test[num_col_names])

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    encoder_col_names = encoder.get_feature_names_out()

    # –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã
    X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
    X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)

    X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
    X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

    # –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    X_train = pd.concat([X_train_scaled, X_train_ohe], axis=1)
    X_test = pd.concat([X_test_scaled, X_test_ohe], axis=1)

    # –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
    model_clf = LogisticRegression()
    model_clf.fit(X_train, y_train)

    # –ü–æ–ª—É—á–µ–Ω–∏–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏
    predictions = model_clf.predict(X_test)
    
    return model_clf, predictions, X_train, X_test, y_train, y_test
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ 
# –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –º–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫

def quality_model_clf(y_test, predictions):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–∏—Ç—ã–≤–∞–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ –¥–ª—è –º–æ–¥–µ–ª–∏
    –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏ –∏ —Å—Ç—Ä–æ–∏—Ç –º–∞—Ç—Ä–∏—Ü—É –æ—à–∏–±–æ–∫
    
    –ù–∞ –≤–≤–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç:
    y_test - —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –¥–ª—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏
    predictions - –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–∞–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏

    –ù–∞ –≤—ã–≤–æ–¥–µ —Ñ—É–Ω—Ü–∏–∏ –º—ã –ø–æ–ª—É—á–∞–µ–º:
    acc - –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ "Accuracy",
    recall - –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ "Recall",
    precision - –∑–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ "Precision",
    matrix_plot - –º–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
    '''

    # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ Accuracy
    acc = accuracy_score(y_test, predictions)

    # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ Recall
    recall = recall_score(y_test, predictions)

    # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ Precision
    precision = precision_score(y_test, predictions)

    # –ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫
    cm = confusion_matrix(y_test, predictions)
    matrix_plot = sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
    plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –æ—à–∏–±–æ–∫', fontsize=14)
    plt.xlabel('Predicted')
    plt.ylabel('True')

    # –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    plt.show()
    print(f'–†–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç—Ä–∏–∫–∏ Accuracy: {acc:.3f}')
    print(f'–†–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç—Ä–∏–∫–∏ Recall: {recall:.3f}')
    print(f'–†–µ–∑—É–ª—å—Ç–∞—Ç –º–µ—Ç—Ä–∏–∫–∏ Precision: {precision:.3f}')

    return acc, recall, precision, matrix_plot, cm
```

```python
def doorstep_clf(low, high, num, data):
    '''
    –î–∞–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å—á–∏—Ç—ã–≤–∞–µ—Ç –º–µ—Ç—Ä–∏–∫—É precision
    —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–æ—Ä–æ–≥–∞–º–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏

    –ù–∞ –≤–≤–æ–¥ —Ñ—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç:
    low - –Ω–∏–∂–Ω–∏–π –ø–æ—Ä–æ–≥ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
    high - –≤–µ—Ä—Ö–Ω–∏–π –ø–æ—Ä–æ–≥ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
    num - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ—Ä–æ–≥–æ–≤
    data - –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ 
    –∏ –∏—Ö —Ä–∞—Å—á–∏—Ç–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏

    –ù–∞ –≤—ã–≤–æ–¥–µ —Ñ—É–Ω–∫—Ü–∏—è –≤—ã–¥–∞–µ—Ç:
    data - –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç—Å—è —Ç–∞ –∂–µ —Ç–∞–±–ª–∏—Ü–∞ —Å –¥–æ–±–∞–≤–ª–µ–Ω–Ω—ã–º–∏ —Å—Ç–æ–ª–±—Ü–∞–º–∏
    –≥–¥–µ —Ä–∞—Å—á–∏—Ç–∞–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–æ—Ä–æ–≥–∞–º–∏
    –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏.
    '''

    
    # –ù–æ–≤—ã–µ –ø–æ—Ä–æ–≥–∏ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
    thresholds = [round(i, 3) for i in np.linspace(low, high, num=num, endpoint=False)]

    # –†–∞—Å—á–∏—Ç—ã–≤–∞–µ–º –Ω–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–æ—Ä–æ–≥–∞–º–∏
    # –°–ø–∏—Å–æ–∫ –¥–ª—è –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫
    columns = []

    # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö —Å—Ç–æ–ª–±—Ü–æ–≤ —Å –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è–º–∏ –≤ —Ç–∞–±–ª–∏—Ü—É
    for i in thresholds:
        columns.append('y_pred_'+str(i))
        data['y_pred_'+str(i)] = data['y_proba'].apply(lambda x: 1 if x>=i else 0)

        # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫–∏ precision
        precision_thresholds = precision_score(data['y_valid'], data['y_pred_'+str(i)])
        print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ Precision c –ø–æ—Ä–æ–≥–æ–º {i}: {precision_thresholds}')

    return data
```

–ù–∞ —á—Ç–æ –º—ã –æ–±—Ä–∞—Ç–∏–ª–∏ –≤–Ω–∏–º–∞–Ω–∏–µ, –∫–æ–≥–¥–∞ –æ–∑–Ω–∞–∫–∞–º–ª–∏–≤–∞–ª–∏—Å—å —Å –æ–±—â–µ–π –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–µ–π –æ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞—Ö:

```python
# –î–µ–ª–∏–º –¥–∞—Ç–∞—Ñ—Ä–µ–π–º –Ω–∞ –≤—Ö–æ–¥–Ω—ã–µ –∏ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫–∏
# –°–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –±—É–¥—É—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è, –∫–∞–∫ –≤—Ö–æ–¥–Ω—ã–µ
drop_list_4 = ['id', 'milk_yield_kg', 
               'efu', 'type_pasture', 
               'spr', 'spr_category', 
               'breed', 'fat_content_%', 
               'protein_content_%', 'taste_milk',
               'daddy_breed', 'efu_9th']

# –í—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
X_4 = df_ferma_main.drop(drop_list_3, axis=1)

# –¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫
y_4 = df_ferma_main['milk_yield_kg']
```

```python
# –°–ø–∏—Å–æ–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
num_col_4 = ['efu_3th', 'crude_protein_g']
cat_col_4 = ['age', 'spr_category_2', 'daddy_name']
```

```python
# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
model_lr_4, predictions_4, X_train_4, X_test_4, y_train_4, y_test_4 = learning_model_lr(
    X=X_4,
    y=y_4,
    size=0.25,
    num_col_names=num_col_4,
    cat_col_names=cat_col_4
)
```

#### –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –∏ –∞–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤:

```python
# –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–∏ –∏ –∞–Ω–∞–ª–∏–∑ –æ—Å—Ç–∞—Ç–∫–æ–≤
r2_4, residuals_4 = quality_model_lr(model_lr=model_lr_4, 
                                     y_test=y_test_4, 
                                     predictions=predictions_4)
```

#### –í—ã–≤–æ–¥:

–î–∞–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –Ω–µ–º–Ω–æ–≥–æ —Ö—É–∂–µ –ø–æ –∫–∞—á–µ—Å—Ç–≤—É –ø–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏—é —Å –ø—Ä–µ–¥—ã–¥—É—â–µ–π –∑–Ω–∞—á–µ–Ω–∏–µ **R2 - 0.867**. –û—Å—Ç–∞—Ç–∫–∏ –±–æ–ª–µ–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –Ω–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–µ, –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è —Å–∏–ª—å–Ω—ã—Ö –æ—Ç–ª–∏—á–∏–π –Ω–µ –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è. –í —Å–ª–µ–¥—É—é—â–µ–º —Ä–∞–∑–¥–µ–ª–µ —Å—Ä–∞–≤–Ω–∏–º –º–æ–¥–µ–ª–∏ –±–æ–ª–µ–µ –ø–æ–¥—Ä–æ–±–Ω–æ. 

### –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –∫–∞—á–µ—Å—Ç–≤–∞ –º–æ–¥–µ–ª–µ–π:

#### –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫ –¥–ª—è –æ—Ü–µ–Ω–∫–∏ –º–æ–¥–µ–ª–µ–π:

```python
# –†–∞—Å—á–µ—Ç MSE
mse_1 = mean_squared_error(y_test_1, predictions_1)
mse_2 = mean_squared_error(y_test_2, predictions_2)
mse_3 = mean_squared_error(y_test_3, predictions_3)
mse_4 = mean_squared_error(y_test_4, predictions_4)

# –†–∞—Å—á–µ—Ç RMSE
rmse_1 = np.sqrt(mse_1)
rmse_2 = np.sqrt(mse_2)
rmse_3 = np.sqrt(mse_3)
rmse_4 = np.sqrt(mse_4)

# –†–∞—Å—á–µ—Ç MAE
mae_1 = mean_absolute_error(y_test_1, predictions_1)
mae_2 = mean_absolute_error(y_test_2, predictions_2)
mae_3 = mean_absolute_error(y_test_3, predictions_3)
mae_4 = mean_absolute_error(y_test_4, predictions_4)
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã —Å –æ—Å—Ç–∞—Ç–∫–∞–º–∏ –º–æ–¥–µ–ª–µ–π
df_quality = pd.DataFrame({'model_1': residuals_1,
                           'model_2': residuals_2,
                           'model_3': residuals_3,
                           'model_4': residuals_4})

# –°–≤–æ–¥–Ω–∞—è —Ç–∞–±–ª–∏—Ü–∞ –ø–æ —Å—Ç–∞—Ç. –º–µ—Ç—Ä–∏–∫–∞–º –æ—Å—Ç–∞—Ç–∫–æ–≤
df_quality_pivot = df_quality.describe().T

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–æ–≤ —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –ø–æ–¥—Å—á–∏—Ç–∞–Ω–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫
df_quality_pivot['r2'] = pd.Series({'model_1': r2_1,
                                    'model_2': r2_2,
                                    'model_3': r2_3,
                                    'model_4': r2_4})

df_quality_pivot['mse'] = pd.Series({'model_1': mse_1,
                                     'model_2': mse_2,
                                     'model_3': mse_3,
                                     'model_4': mse_4})

df_quality_pivot['rmse'] = pd.Series({'model_1': rmse_1,
                                      'model_2': rmse_2,
                                      'model_3': rmse_3,
                                      'model_4': rmse_4})

df_quality_pivot['mae'] = pd.Series({'model_1': mae_1,
                                     'model_2': mae_2,
                                     'model_3': mae_3,
                                     'model_4': mae_4})
```

#### –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –æ—Å—Ç–∞—Ç–∫–æ–≤:

```python
# –°–æ–∑–¥–∞–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# –ì—Ä–∞—Ñ–∏–∫ —è—â–∏–∫–∞ —Å —É—Å–∞–º–∏
sns.boxplot(data=df_quality, ax=axes[0])

# –ì—Ä–∞—Ñ–∏–∫ —è—â–∏–∫–∞ —Å —É—Å–∞–º–∏
sns.violinplot(data=df_quality, ax=axes[1])

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –∑–∞–≥–æ–ª–æ–≤–∫–∞ –∏ –ø–æ–¥–ø–∏—Å–µ–π
fig.suptitle("–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤", fontsize=14)
axes[0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
axes[1].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞')

# –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫
plt.tight_layout()
plt.show()
```

#### –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –º–µ—Ç—Ä–∏–∫:

```python
# –°–ø–∏—Å–æ–∫ —Å—Ç–æ–ª–±—Ü–æ–≤ —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –º–µ—Ç—Ä–∏–∫
model_lr_metrics = ['r2', 'mse', 'rmse', 'mae']
```

```python
# –°–æ–∑–¥–∞–Ω–∏–µ —Å–æ—Å—Ç–∞–≤–Ω–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º axes –≤ –æ–¥–Ω–æ–º–µ—Ä–Ω—ã–π –º–∞—Å—Å–∏–≤
axes = axes.flatten()

# –¶–∏–∫–ª –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
for i, metric in enumerate(model_lr_metrics):
    sns.barplot(data=df_quality_pivot, x=df_quality_pivot.index, y=metric, ax=axes[i])
        
    # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –ø–æ–¥–∑–∞–≥–æ–ª–æ–≤–∫–æ–≤ –∏ –ª–µ–≥–µ–Ω–¥—ã
    axes[i].set_title(f'–°—Ä–∞–≤–Ω–µ–Ω–∏—è –º–æ–¥–µ–ª–µ–π –ø–æ –º–µ—Ç—Ä–∏–∫–µ {metric}')
    axes[i].set_xlabel('')

# –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≥—Ä–∞—Ñ–∏–∫–∏
plt.tight_layout()
plt.show()
df_quality_pivot
```

–ò—Å—Ö–æ–¥—è –∏–∑ –∞–Ω–∞–ª–∏–∑–∞ –≥—Ä–∞—Ñ–∏–∫–æ–≤, –º—ã –≤–∏–¥–∏–º, —á—Ç–æ –ø–æ –≤—Å–µ–º –º–µ—Ç—Ä–∏–∫–∞–º –∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è–º –æ—Å—Ç–∞—Ç–∫–æ–≤ –ª—É—á—à–µ –≤—Å–µ–≥–æ —Å–µ–±—è –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç **–º–æ–¥–µ–ª—å –ø–æ–¥ –Ω–æ–º–µ—Ä–æ–º 3**, –Ω–æ –º—ã –ø–æ–º–Ω–∏–º, —á—Ç–æ —É –Ω–∞—Å –µ—Å—Ç—å –æ–ø–∞—Å–µ–Ω–∏—è, —á—Ç–æ –¥–∞–Ω–Ω–∞—è –º–æ–¥–µ–ª—å –º–æ–∂–µ—Ç –±—ã—Ç—å –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∞ –∏ —Ç–æ–≥–¥–∞ —Å—Ç–æ–∏—Ç –±—É–¥–µ—Ç –≤—ã–±—Ä–∞—Ç—å **–º–æ–¥–µ–ª—å –ø–æ–¥ –Ω–æ–º–µ—Ä–æ–º 4**.

–î–ª—è –æ–∫–æ–Ω—á–∞—Ç–µ–ª—å–Ω–æ–≥–æ –ø–æ–Ω–∏–º–∞–Ω–∏—è –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∞ –ª–∏ **–º–æ–¥–µ–ª—å –ø–æ–¥ –Ω–æ–º–µ—Ä–æ–º 3**, –º—ã —Å—Ä–∞–≤–Ω–∏–º –º–µ—Ç—Ä–∏–∫–∏ —ç—Ç–æ–π –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ —Ç–µ—Å—Ç–æ–≤—ã—Ö, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –ø–æ –¥–∞–Ω–Ω—ã–º –º–µ—Ç—Ä–∏–∫–∞–º –±—É–¥–µ—Ç —Å–∏–ª—å–Ω–æ–π, —Ç–æ –º—ã –æ—Ç–¥–∞–¥–∏–º –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏–µ **4–π –º–æ–¥–µ–ª–∏**, –µ—Å–ª–∏ –∂–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –º–µ—Ç—Ä–∏–∫ –æ–∫–∞–∂—É—Ç—Å—è –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ–π, —Ç–æ –æ—Å—Ç–∞–Ω–æ–≤–∏–º —Å–≤–æ–π –≤—ã–±–æ—Ä –Ω–∞ **–º–æ–¥–µ–ª–∏ –ø–æ–¥ –Ω–æ–º–µ—Ä–æ–º 3**.

#### –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –∏ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–æ–∫ –¥–ª—è —Ç—Ä–µ—Ç—å–µ–π –º–æ–¥–µ–ª–∏:

```python
# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ 3 –º–æ–¥–µ–ª–∏ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
predictions_train_3 = model_lr_3.predict(X_train_3)

# –†–∞—Å—á–µ—Ç –æ—Å—Ç–∞—Ç–∫–æ–≤
residuals_train_3 = y_train_3 - predictions_train_3

# –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫
r2_train_3 = r2_score(y_train_3, predictions_train_3)
mse_train_3 = mean_squared_error(y_train_3, predictions_train_3)
rmse_train_3 = np.sqrt(mse_train_3)
mae_train_3 = mean_absolute_error(y_train_3, predictions_train_3)
```

```python
# –í—ã–≤–æ–¥ –º–µ—Ç—Ä–∏–∫ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ R2 –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {r2_train_3:.3f}')
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ R2 –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {r2_3:.3f}')
print()
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ MSE –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {mse_train_3:.3f}')
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ MSE –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {mse_3:.3f}')
print()
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ RMSE –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {rmse_train_3:.3f}')
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ RMSE –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {rmse_3:.3f}')
print()
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ MAE –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {mae_train_3:.3f}')
print(f'–ó–Ω–∞—á–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫–∏ MAE –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {mae_3:.3f}')
```

–†–∞–∑–Ω–∏—Ü–∞ –≤ –º–µ—Ç—Ä–∏–∫–∞—Ö –º–∏–Ω–∏–º–∞–ª—å–Ω–∞ **R2 –≤—ã—à–µ –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ**, **MSE/RMSE/MAE –≤—ã—Ä–æ—Å–ª–∏ –Ω–∞ 1‚Äì4%**, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ **–æ—Ç—Å—É—Ç—Å—Ç–≤–∏–µ –ø–µ—Ä–µ–æ–±—É—á–µ–Ω–∏—è —Ç—Ä–µ—Ç—å–µ–π –º–æ–¥–µ–ª–∏**.

#### –†–∞—Å—á–µ—Ç –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:

```python
# –†–∞—Å—á–µ—Ç –∫–≤–∞–Ω—Ç–∏–ª–µ–π –æ—à–∏–±–æ–∫
lower_quantile = np.percentile(residuals_train_3, 2.5)
upper_quantile = np.percentile(residuals_train_3, 97.5)

# –†–∞—Å—á–µ—Ç —Ä–∞–∑–º–∞—Ö–∞ –∫–≤–∞–Ω—Ç–∏–ª–µ–π –æ—à–∏–±–æ–∫
range_value = abs(upper_quantile - lower_quantile)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
print('–î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–π –∏–Ω—Ç–µ—Ä–≤–∞–ª —Ç—Ä–µ—Ç—å–µ–π –º–æ–¥–µ–ª–∏:')
print('-'*50)
print(f'–ù–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞: {lower_quantile:.3f}')
print(f'–í–µ—Ä—Ö–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞: {upper_quantile:.3f}')
print(f'–†–∞–∑–º–∞—Ö –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞: {range_value:.3f}')
```

### –ü—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞–Ω–∏–µ —É–¥–æ—è –∫–æ—Ä–æ–≤, —Ä–∞—Å—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º—ã—Ö –¥–ª—è –ø–æ–∫—É–ø–∫–∏:

#### –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º "df_cow_buy":

```python
# –†–∞—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ 
# –¥–ª—è –≤—Å–µ—Ö –∫–æ–º–±–∏–Ω–∞—Ü–∏–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
median_attribute = df_ferma_main.pivot_table(index=['age', 'breed', 
                                                    'type_pasture', 
                                                    'daddy_name'], 
                                             values=['efu', 'crude_protein_g', 'spr'], 
                                             aggfunc='median')
# –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º –∫–æ–ª–æ–Ω–∫–∏
median_attribute.columns = ['crude_protein_g', 'efu', 'spr']

# –£–±–∏—Ä–∞–µ–º –º—É–ª—å—Ç–∏–∏–Ω–¥–µ–∫—Å
median_attribute = median_attribute.reset_index()

# –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º "df_cow_buy"
df_cow_buy = df_cow_buy.merge(median_attribute, 
                              on=['age', 'breed', 
                                  'type_pasture', 
                                  'daddy_name'], 
                              how='left')
```

```python
# –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º –∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ 5%
df_cow_buy['efu'] = df_cow_buy['efu'] * 1.05
df_cow_buy['crude_protein_g'] = df_cow_buy['crude_protein_g'] * 1.05
df_cow_buy['spr'] = df_cow_buy['spr'] * 1.05
```

```python
# –û–∫—Ä—É–≥–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –¥–æ —Ñ–æ—Ä–º–∞—Ç–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
df_cow_buy['crude_protein_g'] = df_cow_buy['crude_protein_g'].astype('int')
df_cow_buy['efu'] = df_cow_buy['efu'].round(1)
df_cow_buy['spr'] = df_cow_buy['spr'].round(3)
```

```python
# –î–æ–±–∞–≤–ª—è–µ–º –ø—Ä–∏–∑–Ω–∞–∫ "efu_9th"
df_cow_buy['efu_9th'] = df_cow_buy['efu']**9
```

```python
# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤—ã–≤–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫ "spr" –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π
df_cow_buy['spr_category_2'] = df_cow_buy['spr'].apply(spr_category_2)
```

```python
# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –¥–æ–±–∞–≤–ª–µ–Ω–∏—è –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
df_cow_buy.head()
```

–ü—Ä–∏ —Ñ–æ—Ä–º–∏—Ä–æ–≤–∞–Ω–∏–∏ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∏—à–ª–æ—Å—å –Ω–µ —É—á–∏—Ç—ã–≤–∞—Ç—å –ø–æ—Ä–æ–¥—É –æ—Ç—Ü–∞, —Ç–∞–∫ –∫–∞–∫ –≤ –¥–∞–Ω–Ω—ã—Ö **"df_ferma_main"** –Ω–µ—Ç –≤—Å–µ—Ö —Å–æ—á–µ—Ç–∞–Ω–∏–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –∏–º–µ—é—Ç—Å—è –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ **"df_cow_buy"**.

#### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö:

```python
# –°–ø–∏—Å–æ–∫ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
input_attribute = ['efu_9th', 'crude_protein_g', 'age', 'spr_category_2', 'daddy_name']
X_buy = df_cow_buy[input_attribute]
```

```python
# –ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
X_buy_ohe = encoder.fit_transform(X_buy[cat_col_3])

# –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
X_buy_scaled = scaler.fit_transform(X_buy[num_col_3])

# –ü–æ–ª—É—á–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
encoder_col_names_buy = encoder.get_feature_names_out()

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã
X_buy_ohe = pd.DataFrame(X_buy_ohe, columns=encoder_col_names_buy)
X_buy_scaled = pd.DataFrame(X_buy_scaled, columns=num_col_3)

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
X_buy = pd.concat([X_buy_scaled, X_buy_ohe], axis=1)
```

```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –ø–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
X_buy.head()
```

–ù–µ—Å–º–æ—Ç—Ä—è –Ω–∞ —Ç–æ, —á—Ç–æ —É –Ω–∞—Å —É–∂–µ –µ—Å—Ç—å –≤ —Ç–∞–±–ª–∏—Ü–µ –∑–Ω–∞—á–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∂–∏—Ä–∞ –∏ –±–µ–ª–∫–∞ –≤ –º–æ–ª–æ–∫–µ –∫–æ—Ä–æ–≤, —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è –¥–∞–Ω—ã –∏—Å—Ö–æ–¥—è —Ç–æ–≥–æ, —á—Ç–æ –∫–æ—Ä–æ–≤–∞–º –¥–∞–≤–∞–ª–∏ –∫–æ—Ä–º –≤ —Ö–æ–∑—è–π—Å—Ç–≤–µ "–≠–∫–æ–§–µ—Ä–º–∞", –Ω–æ –Ω–∞–º –∏–Ω—Ç–µ—Ä–µ—Å–Ω–æ –∫–∞–∫–æ–π –≤–∫—É—Å –º–æ–ª–æ–∫–∞ –±—É–¥–µ—Ç, –µ—Å–ª–∏ –∫–æ—Ä–æ–≤–∞ –±—É–¥–µ—Ç –ø–æ–ª—É—á–∞—Ç—å –∫–æ—Ä–º —É–∂–µ –Ω–∞ —Ñ–µ—Ä–º–µ "–í–æ–ª—å–Ω—ã–π –ª—É–≥". –î–ª—è —ç—Ç–æ–≥–æ –º—ã —Ä–∞—Å—á–∏—Ç–∞–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —ç—Ç–∏—Ö –ø–æ–∫–∞–∑–∞—Ç–µ–ª–µ–π, –∫–∞–∫ –¥–µ–ª–∞–ª–∏ —Ä–∞–Ω–µ–µ –¥–ª—è –¥—Ä—É–≥–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

```python
# –†–∞—Å—á–∏—Ç—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ 
# –¥–ª—è –≤—Å–µ—Ö –∫–æ–º–±–∏–Ω–∞—Ü–∏–π –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
median_attribute_clf = df_ferma_main.pivot_table(index=['age', 'breed', 
                                                        'type_pasture', 
                                                        'daddy_name'], 
                                                 values=['fat_content_%', 'protein_content_%'], 
                                                 aggfunc='median')
# –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤—ã–≤–∞–µ–º –∫–æ–ª–æ–Ω–∫–∏
median_attribute_clf.columns = ['fat_content_%', 'protein_content_%']

# –£–±–∏—Ä–∞–µ–º –º—É–ª—å—Ç–∏–∏–Ω–¥–µ–∫—Å
median_attribute_clf = median_attribute_clf.reset_index()

# –î–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º "df_cow_buy"
df_cow_buy = df_cow_buy.merge(median_attribute_clf, 
                              on=['age', 'breed', 
                                  'type_pasture', 
                                  'daddy_name'], 
                              how='left')
```

```python
# –ü—Ä–∏–∑–Ω–∞–∫ "efu" –≤–æ –≤—Ç–æ—Ä–æ–π —Å—Ç–µ–ø–µ–Ω–∏
df_cow_buy['efu_2th'] = df_cow_buy['efu']**2
```

#### –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö:

```python
# –°–ø–∏—Å–æ–∫ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
input_attribute_clf = ['efu_2th', 'fat_content_%', 'protein_content_%', 'type_pasture', 'spr_category_2']
X_buy_clf = df_cow_buy[input_attribute_clf]
```

```python
# –ö–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
X_buy_ohe = encoder.fit_transform(X_buy_clf[cat_col_clf_2])

# –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
X_buy_scaled = scaler.fit_transform(X_buy_clf[num_col_clf_2])

# –ü–æ–ª—É—á–µ–Ω–∏–µ –Ω–æ–≤—ã—Ö –Ω–∞–∑–≤–∞–Ω–∏–π —Å—Ç–æ–ª–±—Ü–æ–≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
encoder_col_names_buy_clf = encoder.get_feature_names_out()

# –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º—ã
X_buy_ohe = pd.DataFrame(X_buy_ohe, columns=encoder_col_names_buy_clf)
X_buy_scaled = pd.DataFrame(X_buy_scaled, columns=num_col_clf_2)

# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
X_buy_clf = pd.concat([X_buy_scaled, X_buy_ohe], axis=1)
```

```python
# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞ –ø–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
X_buy_clf.head()
```

#### –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –º–æ–¥–µ–ª–∏:

```python
# –°–æ–∑–¥–∞–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –∏ –æ–±—ä–µ–∫—Ç–æ–≤
y_buy_proba = model_clf_2.predict_proba(X_buy_clf)[:, 1]

df_buy = pd.DataFrame(y_buy_proba, columns=['y_proba'])
```

```python
# –í—ã–±—Ä–∞–Ω–Ω—ã–π –ø–æ—Ä–æ–≥ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
thresholds_buy = 0.803

df_buy['y_pred_0.803'] = df_buy['y_proba'].apply(lambda x: 1 if x>=thresholds_buy else 0)
```

```python
# –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤ —Ç–∞–±–ª–∏—Ü—É
df_cow_buy['taste_milk'] = df_buy['y_pred_0.803'].apply(lambda x: '–≤–∫—É—Å–Ω–æ' if x==1 else '–Ω–µ_–≤–∫—É—Å–Ω–æ')
```
